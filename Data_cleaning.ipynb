```python
# import libraries

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# searching duplicated
ds.duplicated().sum()
```

```python
# removing duplicates
ds.drop_duplicates(inplace=True)
```

```python
# checking null value

ds.isnull().sum()
```

```python
# drop emplty cells
ds.dropna(inplace=True)
```

```python
# finding for unique values in year
ds["Airline"].unique()
```

```python
# handling date

ds["Date_of_Journey"] =pd.to_datetime(ds["Date_of_Journey"])
ds.info()
```

```python
print(ds.to_string())
```

```python
def duration_to_mins(x):
    h, m = 0, 0
    if 'h' in x:
        h = int(x.split('h')[0].strip())
        x = x.split('h')[1].strip()
    if 'm' in x:
        m = int(x.split('m')[0].strip())
    return h*60 + m

ds['Duration'] = ds['Duration'].apply(duration_to_mins)
ds
```

```python
# handling Total_Stops
ds['Total_Stops']=ds["Total_Stops"].str.replace("non-stop","0").str.replace(" stops","").str.replace(" stop","")
```

```python
# convert into numeric datatype
ds["Total_Stops"] =pd.to_numeric(ds["Total_Stops"],errors= 'coerce')
```
